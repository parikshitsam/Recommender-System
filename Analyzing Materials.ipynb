```python
cols_to_keep = similar_features + ['Material']
cols_to_keep
```

```python
similarities_df = data[cols_to_keep]
display(len(similarities_df.Material.unique()))
similarities_df.drop_duplicates(inplace = True)
similarities_df[similarities_df.duplicated('Material')].Material.unique()
```

```python
similarities_df
```

```python
features = similar_features
df = similarities_df
features
```

```python
def combine_features(row):
    cf =  row['Brand']+" "+row['Subrand'] + " "+row['SEGMENTS : Pils / Spécialités / Superspécialités/Bouteille Young adult']+" "+row['Variétés'] + " " + row['Segment LE']
    return cf
```

```python
for feature in features:
    df[feature] = df[feature].fillna('')
df["combined_features"] = df.apply(combine_features,axis=1)
```

```python
df
```

```python
cv = CountVectorizer() 
count_matrix = cv.fit_transform(df["combined_features"])
```

```python
count_matrix
```

```python
cosine_sim = cosine_similarity(count_matrix)
```
